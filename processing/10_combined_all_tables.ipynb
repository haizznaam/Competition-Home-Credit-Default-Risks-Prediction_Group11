```python
%%time
# use first features for base joined
feat_path = os.path.join('features', ls_feat_file[0])
df_combined = pd.read_pickle(feat_path, compression='bz2')

# join the remaing features list
for fname in ls_feat_file[1:]:
    feat_path = os.path.join('features', fname)
    df_feat = pd.read_pickle(feat_path, compression='bz2')
    print(fname, df_feat.shape)

    # add table prefix
    tbl_prefix = fname.split('.')[0]
    rename_col = {cname: "{}_{}".format(tbl_prefix, cname) for cname in df_feat.columns if cname != "SK_ID_CURR"}
    df_feat = df_feat.rename(columns = rename_col)

    # merge  
    df_combined = df_combined.merge(df_feat, on = 'SK_ID_CURR', how = 'left')

print("rows, columns", df_combined.shape)
ls_features = [feat for feat in df_combined.columns if feat not in ["SK_ID_CURR"]]
display(df_combined.head())
```
